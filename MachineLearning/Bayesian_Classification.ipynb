# è´å¶æ–¯åˆ†ç±»
è´å¶æ–¯åŒ»ç”Ÿåˆ†è¯Šæ¨¡æ‹Ÿå™¨   
ä½ æ˜¯ä¸€ä½ç—…äººï¼Œä½ èµ°è¿›è¯Šå®¤ï¼Œå‘è´å¶æ–¯åŒ»ç”Ÿæè¿°äº†ä½ çš„ç—‡çŠ¶ï¼šğŸ¤’ â€œæˆ‘å–‰å’™ç—›ï¼Œå¤´å‘æ™•ï¼Œè¿˜æœ‰ç‚¹ä½çƒ§ã€‚â€ åŒ»ç”Ÿè¦åˆ¤æ–­ä½ æ˜¯ä¸æ˜¯å¾—äº†æ„Ÿå†’ã€è¿‡æ•æˆ–æ–°å† ã€‚   
ğŸ‘¨â€âš•ï¸ åŒ»ç”Ÿè¯´ï¼š

- â€œå¦‚æœä½ æ˜¯æ„Ÿå†’ï¼ˆç±»ğ‘1ï¼‰çš„äººï¼Œå‡ºç°è¿™äº›ç—‡çŠ¶çš„æ¦‚ç‡æ˜¯ 80%ã€‚â€

- â€œå¦‚æœä½ æ˜¯è¿‡æ•ï¼ˆç±»ğ‘2â€‹ï¼‰çš„äººï¼Œå‡ºç°è¿™äº›ç—‡çŠ¶çš„æ¦‚ç‡æ˜¯ 30%ã€‚â€

- â€œå¦‚æœä½ æ˜¯æ–°å† ï¼ˆç±»ğ‘3ï¼‰çš„äººï¼Œå‡ºç°è¿™äº›ç—‡çŠ¶çš„æ¦‚ç‡æ˜¯ 90%ã€‚â€

ç„¶åä»–ç»§ç»­æƒ³ï¼šâ€œä¸è¿‡ï¼Œæˆ‘åœ¨æˆ‘ä»¬åŸå¸‚çš„ç»Ÿè®¡æ•°æ®ä¸­çœ‹åˆ°ï¼š

- ä¸€èˆ¬äººå¾—æ„Ÿå†’çš„æ¦‚ç‡æ˜¯ 50%ï¼ˆå…ˆéªŒğ‘ƒ(ğ‘1)ï¼‰

- å¾—è¿‡æ•çš„æ¦‚ç‡æ˜¯ 30%

- å¾—æ–°å† çš„æ¦‚ç‡æ˜¯ 20%â€

```python
import pandas as pd

# ä¸‰ç§ç–¾ç—…ç±»åˆ«
classes = ['æ„Ÿå†’', 'è¿‡æ•', 'æ–°å† ']

# ä¸‰ç§ç–¾ç—…çš„å…ˆéªŒæ¦‚ç‡ P(c)
priors = {
    'æ„Ÿå†’': 0.5,
    'è¿‡æ•': 0.3,
    'æ–°å† ': 0.2
}

# å…¸å‹ç—‡çŠ¶
symptoms = ['å–‰å’™ç—›', 'å¤´æ™•', 'å‘çƒ§']

# æ¯ç§ç–¾ç—…ä¸‹ï¼Œå‡ºç°å„ä¸ªç—‡çŠ¶çš„æ¦‚ç‡ P(x | c)
likelihood = {
    'æ„Ÿå†’': {'å–‰å’™ç—›': 0.8, 'å¤´æ™•': 0.6, 'å‘çƒ§': 0.7},
    'è¿‡æ•': {'å–‰å’™ç—›': 0.2, 'å¤´æ™•': 0.4, 'å‘çƒ§': 0.1},
    'æ–°å† ': {'å–‰å’™ç—›': 0.9, 'å¤´æ™•': 0.8, 'å‘çƒ§': 0.95}
}
```

è®¡ç®—åéªŒæ¦‚ç‡ï¼šP(æ„Ÿå†’âˆ£è¿™äº›ç—‡çŠ¶)=P(è¿™äº›ç—‡çŠ¶âˆ£æ„Ÿå†’)â‹…P(æ„Ÿå†’)   
ç”¨è´å¶æ–¯å…¬å¼ï¼ˆå¿½ç•¥åˆ†æ¯ï¼Œåšç›¸å¯¹å¤§å°æ¯”è¾ƒï¼‰ï¼š

```python
# å‡è®¾ç—…äººå‡ºç°ä»¥ä¸‹ç—‡çŠ¶ï¼š
patient_symptoms = ['å–‰å’™ç—›', 'å¤´æ™•', 'å‘çƒ§']

# ç”¨è´å¶æ–¯å®šç†è®¡ç®—æ¯ä¸ªç–¾ç—…çš„â€œæœªå½’ä¸€åŒ–åéªŒæ¦‚ç‡â€
posteriors = {}
for cls in classes:
    prob = priors[cls]  # P(c)
    for s in patient_symptoms:
        # ä¸ºä»€ä¹ˆè¦è¿ä¹˜ï¼Ÿå› ä¸ºæˆ‘ä»¬è¦è®¡ç®—çš„æ˜¯ï¼šåœ¨æŸä¸ªç±»åˆ« ğ‘ ä¸‹ï¼Œæ‰€æœ‰è¿™äº›ç—‡çŠ¶ï¼ˆç‰¹å¾ï¼‰åŒæ—¶å‘ç”Ÿçš„æ¦‚ç‡ï¼Œ
        prob *= likelihood[cls][s]  # ä¹˜ä¸Š P(xi | c)
    posteriors[cls] = prob
```

```python
# æ„å»ºç»“æœè¡¨æ ¼
posterior_df = pd.DataFrame.from_dict(posteriors, orient='index', columns=['æœªå½’ä¸€åŒ–åéªŒæ¦‚ç‡'])
posterior_df['å½’ä¸€åŒ–åéªŒæ¦‚ç‡'] = posterior_df['æœªå½’ä¸€åŒ–åéªŒæ¦‚ç‡'] / posterior_df['æœªå½’ä¸€åŒ–åéªŒæ¦‚ç‡'].sum()

# æŒ‰ç…§åéªŒæ¦‚ç‡ä»é«˜åˆ°ä½æ’åº
posterior_df = posterior_df.sort_values(by='å½’ä¸€åŒ–åéªŒæ¦‚ç‡', ascending=False)

# æ‰“å°ç»“æœ
print("è´å¶æ–¯è¯Šæ–­ç»“æœï¼š")
posterior_df
```

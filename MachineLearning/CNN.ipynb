# CNN 
## 卷积核与通道

```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
# 模拟 1 张 32x32x3 彩图
# 这里的32x32x3 是图片的尺寸，3是通道数，也就是RGB三个通道
x = torch.randn(1, 3, 32, 32)

# 卷积层：3通道输入，64个卷积核
conv = nn.Conv2d(in_channels=3, out_channels=64, kernel_size=3)
```

```python
# 前向传播
with torch.no_grad():
    out = conv(x)

# 取前 8 个通道画图
fig, axs = plt.subplots(1, 8, figsize=(20, 3))
for i in range(8):
    axs[i].imshow(out[0, i].numpy(), cmap='gray')
    axs[i].axis('off')
    axs[i].set_title(f'Feature {i+1}')
plt.show()
```
